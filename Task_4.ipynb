```python
# Accident Data Analysis (20162023)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load dataset
df = pd.read_csv("/content/accident_sample.csv")

# 2. Ensure 'Year' column exists (if not, extract from Start_Time)
# The original code checked for 'Date' column, but the dataframe has 'Start_Time'
if "Year" not in df.columns and "Start_Time" in df.columns:
    df["Start_Time"] = pd.to_datetime(df["Start_Time"], errors="coerce")
    df["Year"] = df["Start_Time"].dt.year

# 3. Filter data for 20162023
df_filtered = df[(df["Year"] >= 2016) & (df["Year"] <= 2023)]

# 4. Accident counts per year
yearly_counts = df_filtered["Year"].value_counts().sort_index()
print(yearly_counts)

# 5. Plot accidents per year
plt.figure(figsize=(10,6))
sns.barplot(x=yearly_counts.index, y=yearly_counts.values, palette="viridis")
plt.title("Accidents per Year (20162023)")
plt.xlabel("Year")
plt.ylabel("Number of Accidents")
plt.show()

# 6. Optional: Trends by Weather
plt.figure(figsize=(12,6))
sns.countplot(data=df_filtered, x="Year", hue="Weather_Condition") # Using Weather_Condition as identified earlier
plt.title("Accidents by Weather (20162023)")
plt.xlabel("Year")
plt.ylabel("Accident Count")
plt.legend(title="Weather", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# 7. Optional: Trends by Road Condition
plt.figure(figsize=(12,6))
sns.countplot(data=df_filtered, x="Year", hue="Weather_Condition") # Using Weather_Condition as identified earlier
plt.title("Accidents by Road Condition (20162023)")
plt.xlabel("Year")
plt.ylabel("Accident Count")
plt.legend(title="Road Condition", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
